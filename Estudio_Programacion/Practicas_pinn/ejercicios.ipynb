## Numpy libreria de calculo numerico

```python
import numpy as np

# crear un vector x con 5 puntos entre 0 y 1

x = np.linspace(0,1,5)

# elevar al cuadrado cada elemento

y = np.sqrt(x)
print(f'y = {y}')

#suma y promedio

print(f'suma = ',np.sum(y))
print(f'el promedio es = ', np.mean(y))
```
