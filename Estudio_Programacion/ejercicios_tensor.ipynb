## Ejemplo con una sola derivada

```python
import tensorflow as tf

# tf.constant crea un tensor constante, en cambio tf.Variable crea una variable que puede cambiar su valor en un entrenamiento
#x = 3.0
x = tf.constant(3.0)

with tf.GradientTape() as d1:

    d1.watch(x)

    y = tf.sin(x)
    dy_dx = d1.gradient(y, x)

print(dy_dx)
```

## Ejemplo con dos derivadas


```python
import tensorflow as tf

x = tf.Variable(3.0)

with tf.GradientTape(persistent=True) as d2:
    d2.watch(x)
    with tf.GradientTape() as d1:
        d1.watch(x)
        y = x**3+ 2*x

    y_x = d1.gradient(y, x)
y_xx = d2.gradient(y_x, x)

tf.print("y_x:", y_x)
tf.print("y_xx:", y_xx)
```

## EJERCICIO 1

dervida simple

```python
import tensorflow as tf


x = tf.Variable(3.0) # si es tf.variable , no es necesario el t.watch, puesto que identifica quien es la variable, en cambio si fuera constant si es necesario

with tf.GradientTape(persistent = True) as t2:
    with tf.GradientTape() as t1:
        y = 5 * x **2 - 4*x +1 
    y_x = t1.gradient(y,x)

y_xx = t2.gradient(y_x, x)

tf.print("y_x =" , y_x)
tf.print("y_xx =" , y_xx)

```

## Ejercicio 2

```python
import tensorflow as tf

x = tf.Variable(3.0) # tensorflow recibe floats, da error cuando la estructura viene un numero de tipo no flotante

with tf.GradientTape(persistent=True) as t2:
    with tf.GradientTape() as t1:
        y = x**4 + 3 * x **2
    
    y_x = t1.gradient(y,x)
y_xx = t2.gradient(y_x,x)


tf.print(y_x,y_xx)


```
